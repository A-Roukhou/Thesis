# Data Cleaning

```python
#Text cleaning test 1
import nltk #used
from nltk.tokenize import word_tokenize #used
from nltk.corpus import stopwords #used
from nltk.sentiment.vader import SentimentIntensityAnalyzer #used
from nltk.probability import FreqDist #used
from nltk.stem import PorterStemmer #used
import string
from nltk.stem import WordNetLemmatizer
```

```python
#remove basic punctuation from the text without changing the structure of words
all_booking['review_text_1'] = all_booking.review_text.str.replace(",", "")
all_booking['review_text_1'] = all_booking.review_text_1.str.replace(".", "")

#remove the new line sign /n
all_booking['review_text_1'] = all_booking.review_text_1.str.replace('\n', ' ')

#make it all words lower case
all_booking['review_text_1'] = all_booking.review_text_1.apply(lambda x: str(x).lower())
```

```python
#Stopword removal from the dataframe
stop_words = set(open('snowball_expanded.txt'))
all_booking['clean_review_text'] = all_booking['review_text_1'].apply(lambda x: [word for word in str(x).split() if word not in str(stop_words)])
all_booking['clean_review_text']
```

```python
#Obtain hotel names
all_booking['show_all_reviews'] = all_booking.show_all_reviews.str.replace('See all guest reviews for ', '')
all_booking['show_all_reviews']
```

```python
#Stemming: This makes from the words room and rooms for example one word
ps = PorterStemmer()
all_booking['stemmed_clean_review_text'] = all_booking['clean_review_text'].apply(lambda x : [ps.stem(y) for y in x])
all_booking['stemmed_clean_review_text']
#Try Lemmatization as this should provid a better result than stemming
```

# Frequency Distribution

```python
#Remove remaining punctuation from the words
punct = set(string.punctuation)
all_booking['stemmed_clean_review_text_1'] = all_booking.stemmed_clean_review_text.apply(lambda x: ''.join(word for word in str(x) if word not in punct))

#Frequency distribution of the words used in hotels
top_N = 50
a = all_booking['stemmed_clean_review_text_1'].astype(str).str.cat(sep=',')
words = nltk.tokenize.word_tokenize(a)
word_dist = nltk.FreqDist(words)
print (word_dist)

rslt = pd.DataFrame(word_dist.most_common(top_N),
                    columns=['Word', 'Frequency'])
test = rslt.iloc[1:,:]
test
```

```python
#plot of the frequency distribution
import matplotlib.pyplot as plt
# rslt.plot()
fig, ax = plt.subplots(figsize=(13, 8.6))
ax.plot(test.Word, test.Frequency)

ax.set(xlabel='Words', ylabel='Frequency',
       title='Frequency distribution of words used in hotel reviews')
ax.grid()
plt.show()
```

# Sentiment analysis

```python
#Sentiment Analysis starts here
def sentiment_value(paragraph):
    analyser = SentimentIntensityAnalyzer()
    result = analyser.polarity_scores(paragraph)
    score = result['compound']
    return round(score,2)
sample = all_booking['review_text'][10]
print(sample)
print('Sentiment: ' + str(sentiment_value(sample))) 
# print(all_booking)
```

```python
#Create a column with the sentiment scores for each review
all_booking['sentiment_score'] = all_booking.review_text.apply(lambda x: sentiment_value(str(x)))
all_booking.describe()
```
